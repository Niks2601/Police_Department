```python
df['Season'] = df['crime_date'].dt.quarter.map({1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Fall'})
df
```

```python

monthly_category_counts = df.groupby(['month', 'category']).size().unstack(fill_value=0)
monthly_category_counts
```

# Group by month and category to get the count of incidents for each category in each month

```python

# Plotting a line chart to visualize seasonal variations in specific types of crimes
plt.figure(figsize=(12, 8))
sns.lineplot(data=monthly_category_counts, x=monthly_category_counts.index, y='DRUG/NARCOTIC', label='DRUG/NARCOTIC', marker='s')
sns.lineplot(data=monthly_category_counts, x=monthly_category_counts.index, y='ASSAULT', label='ASSAULT', marker='s')
sns.lineplot(data=monthly_category_counts, x=monthly_category_counts.index, y='VANDALISM', label='VANDALISM', marker='s')
sns.lineplot(data=monthly_category_counts, x=monthly_category_counts.index, y='BURGLARY', label='BURGLARY', marker='s')
sns.lineplot(data=monthly_category_counts, x=monthly_category_counts.index, y='VEHICLE THEFT', label='VEHICLE THEFT', marker='s')

plt.title('Seasonal Variations in Specific Types of Crimes')
plt.xlabel('Month')
plt.ylabel('Number of Incidents')

plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.tight_layout()


for line in plt.gca().lines:
    x_data = line.get_xdata()
    y_data = line.get_ydata()
    for i in range(len(x_data)):
        plt.text(x_data[i], y_data[i], f'{y_data[i]:.0f}', color=line.get_color(), ha='center', va='bottom')
        
        
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

# In this we load dataset extract month and season information, and then plot a line chart showing seasonal variations in specific types of crimes DRUG/NARCOTIC, ASSAULT, VANDALISM, BURGLARY and VEHICLE THEFT.

```python
corr_matrix = df.groupby(['category', 'department_district']).size().unstack(fill_value=0).corr()

# Plotting a heatmap to visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation between Crime Categories and Department Districts')
plt.xlabel('Department District')
plt.ylabel('Crime Category')
plt.tight_layout()

plt.show()
```
